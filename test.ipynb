```python
img = cv2.imread('w.jpg', 0)
```

```python
cv2.imshow("Wholesky", img)
cv2.waitKey(0)
```

```python
plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# create a mask
mask = cv2.imread('m2.jpg', 0)
masked_img = cv2.bitwise_and(img, img, mask = mask)
```

```python
plt.imshow(masked_img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
res = cv2.resize(masked_img, None, fx=0.25, fy=0.25, interpolation=cv2.INTER_AREA)
```

```python
cv2.imshow('res', res)
cv2.waitKey(0)
```
