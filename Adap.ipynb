```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('cirrus.jpg')

b,g,r = cv2.split(img)
B = np.array(b)
R = np.array(r)
```

```python
tmp = R-B
tmp.dtype=np.int8
tmp = tmp.astype(np.int16)
tmp = tmp-tmp.min()
tmp = tmp.astype(np.uint8)
```

```python
hist = cv2.calcHist([tmp], [0], None, [256], [0,256])
plt.hist(tmp.ravel(), 256, [0,256])
plt.show()
```

```python
thres, res = cv2.threshold(tmp,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
```

```python
plt.subplot(1,3,1)
plt.imshow(img)
plt.title('img')
plt.xticks([]), plt.yticks([])
plt.show()

plt.subplot(1,3,2)
plt.imshow(tmp, cmap = 'gray')
plt.title('tmp')
plt.xticks([]), plt.yticks([])
plt.show()

plt.subplot(1,3,3)
plt.imshow(res, cmap = 'gray')
plt.title('bin')
plt.xticks([]), plt.yticks([])
plt.show()
```
